```python
for i in range(2010,2024):
    df_i=df[df["统计年度"]==i]
    df_i = df_i[['地区名称', '地区新质生产力水平']]
    df_i.to_csv(f"your_file_{i}.csv", index=False)
    
    
    output_i = """###
    region_data = ["""

    with open(f'your_file_{i}.csv', 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        next(reader)  # 跳过标题行
        for row in reader:
            output_i += f"\n    (\"{row[0]}\", {float(row[1])}),"  # 关键修改点

    output_i += """
        # 请在此处补充你的实际数据...
    ]
    ###"""

    output_i = """###
    region_data = ["""

    with open(f'your_file_{i}.csv', 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        next(reader)  # 跳过标题行
        for row in reader:
            output_i += f"\n    (\"{row[0]}\", {float(row[1])}),"  # 关键修改点

    output_i += """
        # 请在此处补充你的实际数据...
    ]
    ###"""



    # 原始字符串（示例）
    input_str = output_i

    # 提取列表部分并处理注释
    start = input_str.find('[')
    end = input_str.rfind(']') + 1
    list_str = input_str[start:end]

    # 去除注释行
    cleaned_lines = []
    for line in list_str.split('\n'):
        if '#' in line:
            line = line.split('#')[0]  # 移除行内注释
        line = line.strip()
        if line:
            cleaned_lines.append(line)

    # 重新组合成合法Python列表字符串
    cleaned_list_str = '\n'.join(cleaned_lines)

    # 安全转换
    region_data = ast.literal_eval(cleaned_list_str)
    
    c = (
        Map(init_opts=opts.InitOpts(width="1200px", height="800px"))
        .add(
            series_name="新质生产力水平",
            data_pair=region_data,
            maptype="china",
            is_map_symbol_show=False,
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="中国新质生产力水平分布"),
            visualmap_opts=opts.VisualMapOpts(
                min_=0,
                max_=max(v for _, v in region_data),
                range_color=["#f0f9e8", "#0868ac"],  # 颜色渐变
                is_piecewise=False,
                pos_right="5%",    # 向左移动 (原10% → 5%)
                pos_top="30%",    # 向上移动 (默认顶部对齐 → 30%)
                orient="vertical" # 保持垂直方向
            )
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    )

    c.render(f"NQP_map_{i}.html")
```
